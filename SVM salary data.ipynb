```python
import pandas as pd
dftest=pd.read_csv('SalaryData_Test(1).csv')
dftrain=pd.read_csv("SalaryData_Train(1).csv")
dftrain
```

```python
dftrain.drop(["education"],axis=1,inplace=True)
```

```python
from sklearn import preprocessing
le=preprocessing.LabelEncoder()
for column_name in dftrain.columns:
    if dftrain[column_name].dtype == object:
        dftrain[column_name] = le.fit_transform(dftrain[column_name])
    else:
        pass
dftrain
```

```python
X=dftrain.iloc[:10000,:12]
Y=dftrain.iloc[:10000,12]
```

```python
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)
X=norm_func(X)
```

```python
from sklearn import svm
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report


from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.model_selection import train_test_split, cross_val_score
```

```python
param_range = [0.0001, 0.001, 0.01, 0.1, 1.0, 10.0, 100.0, 1000.0]
param_grid = [{'C': param_range,
               'kernel': ['linear']},
               {'C': param_range,
                'gamma': param_range,
                'kernel': ['rbf']}]
svm = SVC(random_state=0)
gs = GridSearchCV(estimator=svm,
                  param_grid=param_grid,
                  scoring='accuracy',
                  cv=10)
gs = gs.fit(X,Y)
```

```python
gs.best_params_ , gs.best_score_ 
```

```python
clf = SVC(C= 100, gamma = 0.1)
clf.fit(X,Y)
```

```python
dftest.drop(["education"],axis=1,inplace=True)
from sklearn import preprocessing
le=preprocessing.LabelEncoder()
for column_name in dftest.columns:
    if dftest[column_name].dtype == object:
        dftest[column_name] = le.fit_transform(dftest[column_name])
    else:
        pass
dftest
```

```python
Xtest=dftest.iloc[:5000,:12]
Ytest=dftest.iloc[:5000,12]
```

```python
Xtest=norm_func(Xtest)
```

```python
y_pred = clf.predict(Xtest)
acc = accuracy_score(Ytest, y_pred) * 100
print("Accuracy =", acc)
confusion_matrix(Ytest, y_pred)
```
